### Creation of dataframe by performing a groupby over age_range and calculation of the count of total samples under each age group.

```python
g = s1.groupby(by='age_range')
g1 = g.count().reset_index()
g1 = g1.drop(["collection_date", "urban_rural", "education", "income", "Vote_Now", "Past_Vote", "CM_satisfaction", "MLA_satisfaction", "age", "gender", "assembly_no", "category", "weight"], 1)
g1
```

```python
for group in labels:
    print("Length of group {}: ".format(group))
    print(len(g.get_group(group)))
```

### Creation of dataframe by performing a group by over age_range and finding the count of total samples for each age group that opted for the JD(U) party in Vote_Now.

```python
s2 = s1[s1['Vote_Now'] == 'JD(U)']
G = s2.groupby(by = ['age_range']).count().reset_index()
G = G.drop(["collection_date", "urban_rural", "education", "income", "response_id", "Past_Vote", "CM_satisfaction", "MLA_satisfaction", "age", "gender", "assembly_no", "category", "weight"], 1)
G
```

### Merge the two dataframes with the common column as age_range.

```python
df = pd.merge(g1, G, on = 'age_range')
df
```

### Pivot Tables

```python
t = s1.groupby(by=['Vote_Now', 'Past_Vote']).count().reset_index()
t = t.drop(["collection_date", "urban_rural", "education", "income", "age_range", "CM_satisfaction", "MLA_satisfaction", "age", "gender", "assembly_no", "category", "weight"], 1)
t.rename(columns = {'response_id' : 'Count'}, inplace=True)
t.head()
```

```python
table = pd.pivot_table(data = t, index = ['Past_Vote'], columns = ['Vote_Now'])
table
```

```python
t1 = s1[["Vote_Now", "Past_Vote", "weight"]]
table1 = pd.pivot_table(data = t1, index = ['Past_Vote'], columns = ['Vote_Now'])
table1
```
